```python
# Task 1 - Data Visualisation

# Needed Modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as matplot

# Load data.csv file
dfRaw = pd.read_csv('data.csv')
```

```python
# a) date month yaer features
dfRaw['date'] = pd.to_datetime(dfRaw['date'], format='%Y-%m-%d')
dfRaw['month'] = dfRaw['date'].dt.month
dfRaw['year'] = dfRaw['date'].dt.year
```

```python
# b) Earthquake Category
tempList = []
for i in dfRaw['magnitude'].tolist():
    tempVar = ''
    if i < 3:
        tempVar = 'Micro'
    elif i < 4:
        tempVar = 'Minor'
    elif i < 5:
        tempVar = 'Light'
    elif i < 6:
        tempVar = 'Moderate'
    elif i < 7:
        tempVar = 'Strong'
    elif i < 8:
        tempVar = 'Major'
    else:
        tempVar = 'Great'
    tempList.append(tempVar)
dfRaw['earthquakeCategory'] = tempList
```

```python
# c) Depth Category
tempList = []
for i in dfRaw['depth'].tolist():
    tempVar = ''
    if i < 71:
        tempVar = 'Shallow'
    elif i < 301:
        tempVar = 'Intermediate'
    else:
        tempVar = 'Deep'
    tempList.append(tempVar)
dfRaw['depthCategory'] = tempList
```

```python
# Answer - Data Visualisation
dfRaw.head()
```

```python
# Visualize using the months
matplot.figure()
sns.countplot(data=dfRaw, x='month')
matplot.xlabel('Month')
matplot.ylabel('Number')
matplot.title('Earthquakes per month')
matplot.show()
```

```python
# Visualize using the years
matplot.figure()
sns.countplot(data=dfRaw, x='year')
matplot.xlabel('Year')
matplot.ylabel('Number')
matplot.title('Earthquakes per year')
matplot.show()
```

```python
# Visualize using the earthquake category
matplot.figure()
sns.countplot(x='earthquakeCategory', data=dfRaw)
matplot.title('Number of earth earthquake based on their categories')
matplot.xlabel('Earthquake Category')
matplot.ylabel('Number')
matplot.show()
```

```python
# Visualize using the depth category
matplot.figure(figsize=(10, 6))
matplot.title('Number of earth earthquake based on their depth categories')
sns.countplot(x='depthCategory', data=dfRaw, palette='viridis')
matplot.xlabel('Depth Category')
matplot.ylabel('Number')
matplot.show()
```
